```python
import pandas as pd

# 1. Data Preparation:
# Load the dataset
df = pd.read_csv('/content/Task1_dataset.csv')
print("Head: ", df.head(),"\n\n")
print("Shape: ",df.shape,"\n\n")
print("Describe: ",df.describe,"\n\n")
```

```python
# Calculate mean, median, and mode
mean_hours = df["Income"].mean()
median_hours = df["Income"].median()
mode_hours = df["Income"].mode()[0]

print(f"Mean Income: {mean_hours}")
print(f"Median Income: {median_hours}")
print(f"Mode Income: {mode_hours}")
```

```python
mean_hours = df["Satisfaction"].mean()
median_hours = df["Satisfaction"].median()
mode_hours = df["Satisfaction"].mode()[0]

print(f"Mean Satisfaction: {mean_hours}")
print(f"Median Satisfaction: {median_hours}")
print(f"Mode Satisfaction: {mode_hours}")
```

```python

mean_hours = df["Age"].mean()
median_hours = df["Age"].median()
mode_hours = df["Age"].mode()[0]

print(f"Mean Age: {mean_hours}")
print(f"Median Age: {median_hours}")
print(f"Mode Age: {mode_hours}")
```
